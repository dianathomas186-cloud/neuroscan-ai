```python
# ============================================================
#   BRAIN TUMOR CLASSIFICATION USING CNN AND TRANSFER LEARNING
#   Framework  : TensorFlow + Keras
#   Architecture: ResNet50 (Transfer Learning + Fine Tuning)
#   Classes    : Glioma, Meningioma, No Tumor, Pituitary
# ============================================================

# ── INSTALL ──────────────────────────────────────────────────
!pip install kagglehub -q

# ── IMPORTS ──────────────────────────────────────────────────
import kagglehub
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Dense, Dropout, GlobalAveragePooling2D,
                                     Conv2D, MaxPooling2D, Flatten, Input)
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import plot_model

from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve)
from sklearn.preprocessing import label_binarize

print("=" * 50)
print("TensorFlow Version :", tf.__version__)
print("GPU Available      :", tf.config.list_physical_devices('GPU'))
print("=" * 50)

# ── STEP 1 : DOWNLOAD DATASET ────────────────────────────────

path = kagglehub.dataset_download("masoudnickparvar/brain-tumor-mri-dataset")
print("Dataset downloaded at:", path)

# ── STEP 2 : SET PATHS ───────────────────────────────────────
train_dir = os.path.join(path, 'Training')
test_dir  = os.path.join(path, 'Testing')

classes = sorted(os.listdir(train_dir))
print("\nClasses found:", classes)
print("\nTraining images per class:")
for cls in classes:
    count = len(os.listdir(os.path.join(train_dir, cls)))
    print(f"  {cls}: {count} images")
print("\nTesting images per class:")
for cls in classes:
    count = len(os.listdir(os.path.join(test_dir, cls)))
    print(f"  {cls}: {count} images")

# ── STEP 3 : VISUALIZE SAMPLE IMAGES ─────────────────────────
# Visualizing one sample MRI image from each class
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
fig.suptitle('Sample MRI Images - One Per Class', fontsize=16, fontweight='bold')
for idx, cls in enumerate(classes):
    cls_path = os.path.join(train_dir, cls)
    img_file = os.listdir(cls_path)[0]
    img = mpimg.imread(os.path.join(cls_path, img_file))
    axes[idx].imshow(img, cmap='gray')
    axes[idx].set_title(cls.upper(), fontsize=12)
    axes[idx].axis('off')
plt.tight_layout()
plt.show()

# ── STEP 4 : PREPROCESSING AND AUGMENTATION ──────────────────

IMG_SIZE   = 224   # ResNet50 input size
BATCH_SIZE = 32    # Hyperparameter: Batch Size
EPOCHS     = 20    # Maximum epochs (EarlyStopping will control)

train_datagen = ImageDataGenerator(
    rescale=1./255,           # Normalize pixel values
    rotation_range=15,        # Augmentation: rotate
    width_shift_range=0.1,    # Augmentation: horizontal shift
    height_shift_range=0.1,   # Augmentation: vertical shift
    horizontal_flip=True,     # Augmentation: flip
    zoom_range=0.1,           # Augmentation: zoom
    validation_split=0.2      # 80/20 split
)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)
val_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

print("\nClass Indices :", train_generator.class_indices)
print("Training samples  :", train_generator.samples)
print("Validation samples:", val_generator.samples)
print("Testing samples   :", test_generator.samples)

# ── STEP 5 : CNN LAYER VISUALIZATION ─────────────────────────

print("\nCNN Architecture Flow:")
print("Input Image (224x224x3)")
print("  → Convolutional Layer (feature extraction using filters)")
print("  → ReLU Activation     (removes negative values)")
print("  → Pooling Layer       (reduces spatial dimensions)")
print("  → Fully Connected     (classification)")
print("  → Softmax Output      (4 class probabilities)")

# ── STEP 6 : BUILD MODEL - RESNET50 TRANSFER LEARNING ────────

base_model = ResNet50(
    weights='imagenet',        # Pretrained ImageNet weights
    include_top=False,         # Exclude original classifier
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)
base_model.trainable = False   # Freeze base layers initially

# Build custom classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)          # Pooling layer
x = Dense(256, activation='relu')(x)     # Hidden layer 1 - ReLU activation
x = Dropout(0.4)(x)                      # Dropout: Hyperparameter = 0.4
x = Dense(128, activation='relu')(x)     # Hidden layer 2 - ReLU activation
x = Dropout(0.3)(x)                      # Dropout: Hyperparameter = 0.3
output = Dense(4, activation='softmax')(x)  # Output layer - Softmax for 4 classes

model = Model(inputs=base_model.input, outputs=output)

# Compile with Adam optimizer and Cross-Entropy loss
model.compile(
    optimizer=Adam(learning_rate=0.0001),     # Hyperparameter: Learning Rate
    loss='categorical_crossentropy',           # Loss Function: Cross-Entropy
    metrics=['accuracy']
)
print("\nModel Summary:")
model.summary()

# ── STEP 7 : CALLBACKS (HYPERPARAMETER TUNING) ───────────────

callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        verbose=1
    ),
    ModelCheckpoint(
        'best_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=3,
        min_lr=1e-7,
        verbose=1
    )
]

# ── STEP 8 : TRAIN PHASE 1 (FROZEN BASE) ─────────────────────

print("\nPhase 1 Training - Frozen ResNet50 base...")
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    callbacks=callbacks,
    verbose=1
)
print("Phase 1 Training Complete!")

# ── STEP 9 : FINE TUNING (UNFREEZE LAST 20 LAYERS) ───────────

print("\nPhase 2 - Fine Tuning last 20 layers of ResNet50...")
for layer in base_model.layers[-20:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(learning_rate=0.00001),   # Very low LR for fine tuning
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_finetune = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator,
    callbacks=callbacks,
    verbose=1
)
print("Fine Tuning Complete!")

# ── STEP 10 : PLOT ACCURACY AND LOSS CURVES ──────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Brain Tumor Classification - Learning Curves', fontsize=14, fontweight='bold')

axes[0].plot(history.history['accuracy'], label='Train Accuracy (Phase1)', color='blue')
axes[0].plot(history.history['val_accuracy'], label='Val Accuracy (Phase1)', color='orange')
axes[0].plot(history_finetune.history['accuracy'], label='Train Accuracy (FineTune)', color='green')
axes[0].plot(history_finetune.history['val_accuracy'], label='Val Accuracy (FineTune)', color='red')
axes[0].set_title('Model Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history.history['loss'], label='Train Loss (Phase1)', color='blue')
axes[1].plot(history.history['val_loss'], label='Val Loss (Phase1)', color='orange')
axes[1].plot(history_finetune.history['loss'], label='Train Loss (FineTune)', color='green')
axes[1].plot(history_finetune.history['val_loss'], label='Val Loss (FineTune)', color='red')
axes[1].set_title('Model Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# ── STEP 11 : EVALUATE ON TEST DATA ──────────────────────────
print("\nEvaluating on Test Data...")
test_loss, test_accuracy = model.evaluate(test_generator, verbose=1)
print(f"\nTest Accuracy : {test_accuracy*100:.2f}%")
print(f"Test Loss     : {test_loss:.4f}")

# ── STEP 12 : CONFUSION MATRIX ───────────────────────────────
y_pred        = model.predict(test_generator, verbose=1)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true        = test_generator.classes
class_names   = list(test_generator.class_indices.keys())

cm = confusion_matrix(y_true, y_pred_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names,
            yticklabels=class_names)
plt.title('Confusion Matrix - Brain Tumor Classification', fontweight='bold')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

# ── STEP 13 : CLASSIFICATION REPORT ──────────────────────────
print("\nClassification Report (Accuracy, F1-Score, Precision, Recall):")
print(classification_report(y_true, y_pred_classes, target_names=class_names))

# ── STEP 14 : ROC-AUC CURVE ──────────────────────────────────
# Multi-class ROC-AUC as required by the document
y_true_bin = label_binarize(y_true, classes=[0, 1, 2, 3])

plt.figure(figsize=(10, 6))
colors = ['blue', 'orange', 'green', 'red']
for i, (cls, color) in enumerate(zip(class_names, colors)):
    fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_pred[:, i])
    auc_score   = roc_auc_score(y_true_bin[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, color=color, label=f'{cls} (AUC = {auc_score:.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.title('ROC-AUC Curve - Brain Tumor Classification', fontweight='bold')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.tight_layout()
plt.show()

overall_auc = roc_auc_score(y_true_bin, y_pred, average='macro')
print(f"\nOverall ROC-AUC Score (Macro): {overall_auc:.4f}")

# ── STEP 15 : ERROR ANALYSIS ──────────────────────────────────
# Show wrongly predicted MRI images for error analysis
print("\nError Analysis - Wrongly Predicted Images:")
errors        = np.where(y_pred_classes != y_true)[0]
print(f"Total wrong predictions: {len(errors)} out of {len(y_true)}")

fig, axes = plt.subplots(2, 5, figsize=(20, 8))
fig.suptitle('Error Analysis - Misclassified MRI Images', fontsize=14, fontweight='bold')
axes = axes.flatten()

test_image_paths = []
for cls in class_names:
    cls_path = os.path.join(test_dir, cls)
    for img_file in os.listdir(cls_path):
        test_image_paths.append(os.path.join(cls_path, img_file))

for i, err_idx in enumerate(errors[:10]):
    if i >= 10:
        break
    img_path = test_image_paths[err_idx]
    img      = load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
    axes[i].imshow(img)
    axes[i].set_title(
        f"True: {class_names[y_true[err_idx]]}\nPred: {class_names[y_pred_classes[err_idx]]}",
        fontsize=9, color='red'
    )
    axes[i].axis('off')

plt.tight_layout()
plt.show()

# ── STEP 16 : SAVE AND DOWNLOAD MODEL ────────────────────────
model.save('brain_tumor_classification_model.h5')
print("\nModel saved as brain_tumor_classification_model.h5")

from google.colab import files
files.download('brain_tumor_classification_model.h5')
print("Model downloaded to your laptop successfully!")

print("\n" + "=" * 50)
print("PROJECT COMPLETE!")
print(f"Final Test Accuracy : {test_accuracy*100:.2f}%")
print(f"Overall ROC-AUC     : {overall_auc:.4f}")
print("=" * 50)
```
